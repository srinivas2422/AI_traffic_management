```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load metrics
metrics = pd.read_csv("../outputs/metrics.csv")

# Step 2: Plot vehicle counts over frames
plt.figure(figsize=(10, 6))
sns.lineplot(data=metrics, x=metrics.index, y="Vehicle_Count")
plt.title("Vehicle Count Over Frames")
plt.xlabel("Frame Index")
plt.ylabel("Vehicle Count")
plt.show()

# Step 3: PSO Optimization Results Visualization
iterations = list(range(1, 51))  # Example for 50 iterations
fitness_scores = [100 / i for i in iterations]  # Simulated fitness scores

plt.figure(figsize=(10, 6))
plt.plot(iterations, fitness_scores, marker='o', label="Fitness Score")
plt.title("PSO Fitness Score Over Iterations")
plt.xlabel("Iteration")
plt.ylabel("Fitness Score")
plt.legend()
plt.grid()
plt.show()
```
